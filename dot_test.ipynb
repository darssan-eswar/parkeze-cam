```python
from PIL import ImageDraw

def draw_dots(image, top_coords, bottom_coords, width=12, height=28, radius=3):
    """
    Draw translucent vertically elongated rounded rectangles with soft corners.

    Args:
        image: PIL Image object (RGBA)
        top_coords: list of (x,y) tuples for top spots (green)
        bottom_coords: list of (x,y) tuples for bottom spots (red)
        width: rectangle width
        height: rectangle height
        radius: corner radius (small for slight rounding)
    Returns:
        Image with rectangles drawn
    """
    img_copy = image.copy()
    draw = ImageDraw.Draw(img_copy, 'RGBA')

    green_fill = (0, 180, 0, 100)   # softer green, more translucent
    red_fill = (180, 0, 0, 100)     # softer red, more translucent

    for x, y in top_coords:
        rect = [x - width//2, y - height//2, x + width//2, y + height//2]
        draw.rounded_rectangle(rect, radius=radius, fill=green_fill)

    for x, y in bottom_coords:
        rect = [x - width//2, y - height//2, x + width//2, y + height//2]
        draw.rounded_rectangle(rect, radius=radius, fill=red_fill)

    return img_copy
```

```python
# Edit these coordinates to adjust dot positions
top_spots = [(100, 100), (145, 100)]                # 2 dots on top
bottom_spots = [(130, 490), (163, 490), (195, 489)]  # 3 dots on bottom


img_with_dots = draw_dots(lot_map, top_spots, bottom_spots)

plt.figure(figsize=(10, 8))
plt.imshow(img_with_dots)
plt.axis('off')
plt.title("Adjust dot positions by editing 'top_spots' and 'bottom_spots'")
plt.show()
```
